```python
import pandas as pd
import numpy as np
Auto = pd.read_csv('https://www.statlearning.com/s/Auto.data',
                    na_values=['?'],
                    sep=r'\s+')

Auto
```

### 9a

Alla är kvantitativa utom modellnamnen, möjligen med undantag för __origin__ som vi inte har någon förklaring till.

```python
#9b

for j in Auto.columns:
    if j!="name":
        print(f"The range of {j} is {Auto[j].min()} to {Auto[j].max()}, which is {Auto[j].max()-Auto[j].min()}.")
```

```python
#9c

for j in Auto.columns:
    if j!="name":
        print(f"The mean of {j} is {np.mean(Auto[j])} and the standard deviation is {np.std(Auto[j])}.")
```

```python
#9d

Auto_d = Auto.drop(Auto.loc[10:85].index)

for j in Auto_d.columns:
    if j!="name":
        print(f"The range of {j} is {Auto_d[j].min()} to {Auto_d[j].max()}, which is {Auto_d[j].max()-Auto_d[j].min()}.")

for j in Auto_d.columns:
    if j!="name":
        print(f"The mean of {j} is {np.mean(Auto_d[j])} and the standard deviation is {np.std(Auto_d[j])}.")

```

```python
# 9e

from matplotlib.pyplot import subplots


fig,axes=subplots(nrows=2, ncols=3, figsize=(12,7))
axes[0,0].plot(Auto["weight"],Auto["year"], "o");
axes[0,1].scatter(Auto["weight"],Auto["mpg"],marker="o",color="red");
axes[0,2].hist(Auto["weight"], bins=5);
axes[1,0].boxplot(Auto['mpg']);
Auto.boxplot('weight', by='cylinders', ax=axes[1,1]);
Auto.boxplot('mpg', by='origin', ax=axes[1,2]);

```

### 9f

_Weight_ (vikt) och _displacement_ (cylindervolym) verkar vara omvänt proportionella mot _mpg_ (miles per gallon), vilket låter rimligt.
