```python
rng=np.random.default_rng(729)
x=rng.normal(loc=100, scale=2, size=10)
y=np.random.normal(60,5,10)

print(x)
print(y)
np.corrcoef(x,y)

print(np.var(x), np.std(x), np.mean(x))



```

```python
from matplotlib.pyplot import subplots
fig, ax = subplots(figsize=(5, 5))
x=rng.standard_normal(100)
y=rng.standard_normal(100)
ax.scatter(x,y,marker="o")
ax.set_xlabel("x-axeln")
ax.set_ylabel("y-axeln")
ax.set_title("Titel")
fig.set_size_inches(6,4)

fig,axes=subplots(nrows=2, ncols=3, figsize=(12,4));
axes[0,1].plot(x,y, "o")
axes[1,2].scatter(x+1,y,marker="+")

fig.savefig("Figure.png", dpi=400)
fig.savefig("Figure.pdf", dpi=200)

axes[1,1].set_xlim([-10,10])
fig.savefig("Figure.jpg", dpi=400)


```

```python
fig,ax=subplots(figsize=(6,6))
x=np.linspace(-np.pi,np.pi,50)
y=x
f=np.multiply.outer(np.cos(y),1/(1+x**2))
ax.contour(x,y,f,levels=45);
ax.imshow(f);

```
